```
import numpy as np
import random
import math
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
```

```
img = Image.open('/content/sample_img1.png')
img_arr=np.array(img)
print(type(img_arr),img_arr.shape)
plt.figure()
plt.imshow(img_arr)
plt.show()
```

```
class kmeans:
  def __init__(self,k,x):
    if x.shape[-1] == 2:
      self.centers=np.ones((k,1,2))
    else:
      self.centers=np.ones((k,1))
    
  def fit(self,x):
    for i in range(self.centers.shape[0]):
        rand =random.randint(0,1)
        # self.centers[i] = rand * self.centers[i]
        self.centers[i] = x[rand]

    # print("Initial Values for Centers : ")
    # print( self.centers)
    for _ in range(15):
      self.d = np.ones((x.shape[0],self.centers.shape[0]))
      for j in range(self.centers.shape[0]):
        for i in range(x.shape[0]):
          # print((x[i]-self.centers[j]).shape)
          # print((x[i]-self.centers[j]).shape)
          self.d[i,j]= np.linalg.norm((x[i]-self.centers[j]))
          
          # self.d[i,j]=2
      # d[0]=x[i]-self.centers[0]
      # d[1]=x[i]-self.centers[0]
      # if math.abs
      # print(d)
      
      self.labels=np.ones(x.shape[0])
      for i in range(x.shape[0]):
        # print(d[: ,i])
        # print(d)
        self.labels[i]=np.argmin(self.d[i])
        # if abs(d[0,i])>abs(d[1,i]):
        #   self.labels[i]=1
        # else:
        #   self.labels[i]=0

      for i in range(self.centers.shape[0]):
        num = 0
        sum = 0
        # print(i)
        for j in range(x.shape[0]):
          if self.labels[j] == i:
            num += 1
            sum += x[j]
          else :
            pass
        # print(num)
        # print(sum)
        if num != 0:
          self.centers[i] = sum/num
        # print(self.centers)
        
        # if(m==0):
        #   m=1
        
      #   # print(c/m)
  def error_cal(self,x):
    e = 0 
    iter = 0
    for i in range(x.shape[0]):
      iter += 1
      e += np.linalg.norm(self.centers[int(self.labels[i])]-x[i])
    return e/iter


def replace(x,z):
  for i in range(x.shape[0]):
    a=z.labels[i]
    # print(int(a))
    x[i]=z.centers[int(a)]
```

```
one_dim_1=np.reshape(img_arr[:,:,0],(img_arr.shape[0]*img_arr.shape[1],1))
one_dim_1 = one_dim_1.astype('float64') 
one_dim_2=np.reshape(img_arr[:,:,1],(img_arr.shape[0]*img_arr.shape[1],1))
one_dim_2 = one_dim_2.astype('float64') 
one_dim_3=np.reshape(img_arr[:,:,2],(img_arr.shape[0]*img_arr.shape[1],1))
one_dim_3 = one_dim_3.astype('float64') 
```

```
k2 = kmeans(2,one_dim_1)
k2.fit(one_dim_1)
```

```
k2_ = kmeans(1,one_dim_2)
k2_.fit(one_dim_2)
```

```
k2__ = kmeans(1,one_dim_3)
k2__.fit(one_dim_3)
```

```
replace(one_dim_1,k2)
replace(one_dim_2,k2_)
replace(one_dim_3,k2__)
```
